```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('a.jpg') 
print("Original Shape:", img.shape)  
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()
bgr_img = img[..., ::-1]
plt.imshow(bgr_img)
plt.title("BGR Image (Reversed RGB)")
plt.axis('off')
plt.show()
grayscale = 0.299 * img[..., 0] + 0.587 * img[..., 1] + 0.114 * img[..., 2]
grayscale = grayscale.astype(np.uint8)
plt.imshow(grayscale, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
r_df = pd.DataFrame(img[..., 0])
g_df = pd.DataFrame(img[..., 1])
b_df = pd.DataFrame(img[..., 2])
print("Red Channel DataFrame:\n", r_df.head())
print("Green Channel DataFrame:\n", g_df.head())
print("Blue Channel DataFrame:\n", b_df.head())
img_mod = img.copy()
img_mod[..., 0] = 1 
plt.imshow(img_mod)
plt.title("Modified Image (Red Max)")
plt.axis('off')
plt.show()
```
